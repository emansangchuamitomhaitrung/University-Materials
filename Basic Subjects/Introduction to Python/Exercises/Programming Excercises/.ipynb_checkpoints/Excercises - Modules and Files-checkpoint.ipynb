### Excercise 2
Write a function to convert the numpy matrix into percentage matrix: each element in a line is converted into its percentage of the sum of elements in that line (no need for rounding)


```python
def convert(a):
    return a/np.sum(a, axis=1, keepdims=True)

a = [[0, 1, 2],
 [3, 4, 5],
 [6, 7, 8]]

print(convert(a))
```

### Excercise 3
Solve the system of linear equation Ax = b, knowing the augmented matrix [A|b]. It is guarantee that the system of linear equation has n equations, n variables and a single solution

```python
import numpy as np
def system_solver(a):
    b = [row[-1] for row in a]
    a = np.delete(a, -1, axis = 1)
    return np.linalg.solve(a, b).reshape(-1, 1)
```

```python
a = [[1, 3, -2, 5],
     [3 ,5, 6 ,7],
     [2, 4, 3, 8]]

print(system_solver(a))
```

### Question 4
Plot the function y = sin(2x) and y = cos(2x) in the range [-4, 4] on the same plot as below using matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-4, 4, 0.001)

y_sin = np.sin(2 * x)
y_cos = np.cos(2 * x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$']) # changing pi-representation
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of trigonometric function')
plt.legend(['y = sin(2x)', 'y = cos(2x)'], loc='upper right')
plt.grid()
plt.show()
```

### Question 5
Write a module that contains the constant pi and implementations of these mathematics function: exp, sin, cos using Maclaurin series. The result should be computed with 9 exact decimal places and rounded to 6 digits after the comma. Note: you should implement your own function, and do not use function from other libraries

The module will be used for testing later

```python
pi = 3.14159265358979
def exp(x):
    result = 1
    a = x
    i = 1
    while a > 1e-9 or a < -1e-9:
        result += a
        i += 1
        a = a*x/i
    return round(result, 6)

def sin(x):
    x = x - 2*pi*int(x/(2*pi))
    i = 1
    result = 0
    a = x
    while a > 1e-9 or a < -1e-9:
        result += a
        i += 2
        a = -a*x*x/(i-1)/i
    return round(result, 6)

def cos(x):
    return sin(x + pi/2)
```

### Excercise 7
Write a program to count the frequency of words from test.inp (already uploaded to the system) and write the string representation of the dictionary of frequency to the file test.out


```python
import json
f = open('test.inp')
d = dict()
for lines in f:
    for word in lines.split(' '):
        if word in d.keys():
            d[word] += 1
        else:
            d[word] = 1
print(d)
with open('test.out', 'w') as f:
    f.write(json.dumps(d))
```

### Exercise 8
Read 2 same-length lists of numbers and write both lists into a file using module pickle. Then load the 2 lists from that file and print the result list which is the index-wise summation of 2 lists

```python
import pickle
a = [int(i) for i in input().split()]
b = [int(i) for i in input().split()]

with open('lists.pkl', 'wb') as fd:
    pickle.dump([a, b], fd)
with open('lists.pkl', 'rb') as f:
    c, d = pickle.load(f)
result = [i + j for i, j in zip(c, d)]
print(result)
```

### Excercise 9

student_info.pkl contains a dumped list of students' information (already uploaded to the system). We create a copy of that file, named updated_info.pkl, to update new students.
Write a function to add the information of a student to the dumped list in updated_info.pkl. If the student ID in that information piece has already existed in the list, overwrite all of that student's information with the new one. If the student ID is missing, do nothing

```python
student_info = [
  {
   'id': 20200000,
   'name': 'Nguyen Van A',
   'score': {
    'math': 7.8,
    'english': 8.9,
    'physics': 9.0,
   }
  },
  {
   'id': 20200001,
   'name': 'Le Van B',
   'score': {
    'math': 9.8,
    'english': 8.7,
    'physics': 7.6,
   }
  }
]
```

```python
from shutil import copyfile

copyfile('student_info.pkl', 'updated_info.pkl')


def add_student_info(new_student):
    if 'id' not in new_student:
        return

    update_lst = []
    with open('updated_info.pkl', 'rb') as f:
        update_lst = pickle.load(f)

    for i in range(len(update_lst)):
        if new_student['id'] == update_lst[i]['id']:
            update_lst[i] = new_student
            break
    else:
        update_lst.append(new_student)

    with open('updated_info.pkl', 'wb') as f:
        pickle.dump(update_lst, f)
```
